```python
df.replace("..", pd.NA, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Selecting only year columns
year_columns = df.columns[4:]
df[year_columns] = df[year_columns].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

```python
df[year_columns] = df.groupby(["Country Name", "Series Name"])[year_columns].ffill()
df[year_columns] = df.groupby(["Country Name", "Series Name"])[year_columns].bfill()
```

```python
df.isnull().sum()
```

```python
missing_threshold = 0.5 * len(year_columns)
df = df.dropna(thresh=missing_threshold, axis=0)
```

```python
df.shape
```

```python
df[year_columns] = df[year_columns].apply(lambda x: x.fillna(x.mean()), axis=0)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.columns = [col.replace(" [YR", "_").replace("]", "") if "YR" in col else col for col in df.columns]
```

```python
df.reset_index(drop=True, inplace=True)
```
